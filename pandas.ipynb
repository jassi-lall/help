# pandas
`pandas` is a powerful and popular python module for manipulating and understanding tabular data. <br><br>
The central objects in pandas are the `DataFrame` and the `Series`. A DataFrame can be thought of as a table with columns and records (or as a dict like container for series objects). A Series can be thought of as a column of a DataFrame (or a one-dimensional ndarray with axis labels).

## Create a DataFrame
There are a few ways:
- use a csv or excel file: `pd.read_csv(filepath, ...)` or `read_excel`. The function is well endowed with *many* important optional [arguments](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html).
- Note that pandas has so much more functionality. Including reading html tables, and running sql queries on a DataFrame (pd.read_sql)
- use constructor: `pd.DataFrame(data, index, columns, dtype, copy)`.

## Accessing elements & arithmetic of Series
Use `.loc` and  `.iloc` for accessing records and columns. Use `.at` and `iat` for accessing single entries. These are special objects that can be indexed using standard python notation. df.loc[record]
