## Example

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(3)

N = 1000
data = np.empty((N,), dtype=np.int64)

for i in range(N):
    steps = 5
    move = np.random.randint(-1,2,steps)
    end = move.sum()
    data[i] = end

df = pd.DataFrame(data)
s = df.value_counts()
s = s.reset_index()
s.columns = ['end_pos','frequency']
sns.barplot(s, x='end_pos', y='frequency')
plt.show()
````
