# Data visualisation
One can use python (seaborn, matplotlib.pyplot, & pandas) as well as a tool like Excel to create beautiful visualisations that tell a story.

```python
import pandas as pd                 # Data handling - e.g. import csv/excel files, etc
import matplotlib.pyplot as plt     # Data visualisation
import seaborn as sns               # Built on top of matplotlib.pyplot . Makes plots easy and elegant.
```

```python
plt.title('my_title')
plt.ylabel('y')
plt.xlabel('x')
# ..
# ..
sns.lineplot(x=[0,1,2,3,4,5,6,7], y=[2,3,5,7,9,11,13,17])
# Data can be dataframe, numpy ndarray and so on.
# Many examples such as histplot, boxplot, barplot, heatmap
# Use hue optional arg to generate beautiful visuals /w DataFrames that show different categories. Can change legend positioning, names, etc using plt.legend(..)
plt.show()
```
